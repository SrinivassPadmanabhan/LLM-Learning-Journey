```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, AutoModel, AutoModelForMaskedLM, utils
import torch
import os
import dotenv
dotenv.load_dotenv(dotenv_path=r"../../.env")
```

```python
hf_token = os.getenv("hugging_face_token") 
model_name = "bert-base-uncased"
utils.logging.set_verbosity_error()
```

```python
tokeniser = AutoTokenizer.from_pretrained(model_name, token=hf_token)
model = AutoModel.from_pretrained(model_name,token=hf_token)
```

```python
sentence1 = "She took a rose from the garden."
sentence2 = "The sun rose in the morning."
```

```python
input_ids1 = tokeniser(sentence1, return_tensors="pt")["input_ids"]
input_ids2 = tokeniser(sentence2, return_tensors="pt")["input_ids"]
```

```python
for i in range(input_ids1.shape[1]):
    print(input_ids1[0][i].item(), tokeniser.decode(input_ids1[0][i].item()))
```

```python
for i in range(input_ids2.shape[1]):
    print(input_ids2[0][i].item(), tokeniser.decode(input_ids2[0][i].item()))
```

```python
model.eval()
with torch.no_grad():
    output1 = model(input_ids1)
    output2 = model(input_ids2)
```

```python
output1.last_hidden_state[0][5]
```

```python
emb1 = output1.last_hidden_state[0][5]
emb2 = output2.last_hidden_state[0][4]
```

```python
torch.cosine_similarity(emb1, emb2, dim=0)
```
